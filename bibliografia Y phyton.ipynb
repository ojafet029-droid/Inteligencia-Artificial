### 6-.Control de flujo (if, for, while)

##### EJEMPLO

```python
# if
edad = 18
if edad >= 18:
    print("Eres mayor de edad")

# for
for i in range(1, 6):
    print("NÃºmero:", i)

# while
contador = 3
while contador > 0:
    print("Cuenta regresiva:", contador)
    contador -= 1
```

### 8-.Compresiones

##### EJEMPLO

```python
cuadrados = [x*x for x in range(1, 6)]
pares = [x for x in range(10) if x%2==0]
dicc = {x: x**2 for x in range(1, 6)}

print(cuadrados)  # [1, 4, 9, 16, 25]
print(pares)      # [0, 2, 4, 6, 8]
print(dicc)       # {1:1, 2:4, 3:9, 4:16, 5:25}
```

<span style="color:red;">Esto es como un atajo para crear listas o colecciones rÃ¡pido y sin tener que escribir tanto. ### 12-. EXPECIONES

##### EJEMPLO

```python

def dividir_seguro(a, b):
    if b == 0:
        raise ValueError("No se puede dividir entre cero.")
    return a / b

# Ejemplos con manejo de excepciones
casos = [(10, 2), (5, 0), ("9", 3)]
for x, y in casos:
    try:
        resultado = dividir_seguro(x, y)
    except TypeError as e:
        print(f"TypeError con ({x}, {y}): {e}")
    except ValueError as e:
        print(f"ValueError con ({x}, {y}): {e}")
    else:
        print(f"{x} / {y} = {resultado}")
    finally:
        pass  # lugar para liberar recursos si fuera necesario
```

<span style="color:PINK;">AquÃ­ se trata de cÃ³mo manejar los errores que pasan en un programa. Muchas veces algo puede fallar, como dividir entre cero, abrir un archivo que no existe o escribir mal un dato. Si no hago nada, el programa se detiene de golpe. Pero con las excepciones, en lugar de que se cierre, puedo darle una respuesta, como mostrar un mensaje de error o seguir con otra parte del cÃ³digo. Esto hace que mi programa sea mÃ¡s seguro y no se caiga tan fÃ¡cil.</SPAM>
